```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
url = "https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet"
taxi_trip = pd.read_parquet(
    url, 
    engine="pyarrow"
)
```

```python
url2 = "https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/taxi_zone_lookup.csv"
taxi_zone = pd.read_csv(url2)
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
taxi_trip.to_sql(name='taxi_trip_201511', con=engine, if_exists='replace')
```

```python
taxi_zone.to_sql(name='taxi_zone', con=engine, if_exists='replace')
```
